```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--start-maximized")
# Define the Chrome webdriver options
driver = webdriver.Chrome(options=chrome_options)
# driver = webdriver.Firefox()

# driver.get("https://resultados.gob.ar/elecciones/1/73994/1/-1/-1/C%C3%B3rdoba/Capital/00001/ESC--NACIONAL-DE-MONSERRAT/0400100009X")
table_map = {}
table_counter = 23954 
max_counter = 0
not_found_pages_list = []
not_data_page_list = []

#while table_counter < 23954+10 :
for table_counter in range(23954,23954+10):
    url = f"https://resultados.gob.ar/elecciones/1/{table_counter}/1/-1/-1/"
    driver.get(url)
    party_map = {}
    
    try:
        # Espera a que un elemento con el ID "element_id" esté presente en la página
        first_element = WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.XPATH, '//div[@class="flex flex-col min-h-screen"]//p[@class="font-bold text-[#3E3D3D] text-[10px] md:text-sm leading-1 md:leading-3"]'))
        )    
        table_id = first_element.text
        second_element = WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.XPATH, '//div[@class="flex flex-col xl:flex-row"]//div[@class="flex-1 lg:max-h-[900px] lg:max-w-[1000px]"]//ul'))
            #EC.presence_of_element_located((By.XPATH, '//div[@class="flex-1 lg:max-h-[900px] lg:max-w-[1000px]"]//ul[@class="p-4 md:p-0 flex flex-col gap-4"]'))
        )    
        # Obtener todos los elementos "li" dentro de "elemento"
        li_elements = second_element.find_elements(By.TAG_NAME, 'li')

        # Iterar a través de los elementos "li"
        for li in li_elements:
            # Encontrar el elemento "p" dentro de cada elemento "li"
            party_element = li.find_element(By.TAG_NAME, 'p')
            votes_element = li.find_element(By.TAG_NAME, 'span')
            # Obtener el texto dentro del elemento "p"
            party = party_element.text
            votes = votes_element.text
            # Imprimir o procesar el texto según sea necesario
            party_map[party] = votes

        table_map[table_id] = party_map 
        #table_counter +=1

        """if ((len(party_map.values()) == 1) and (list(party_map.values())[0] == '')):
            table_counter-=1    
            max_counter +=1    
            print(party_map)    
        if max_counter==3:
            max_counter=0
            table_counter +=1
            not_data_page_list.append(table_counter)
        print(table_counter)
"""
    except Exception:
            # En caso de que expire el tiempo de espera
            print( f"Elemento: {table_counter} no encontrado en la página.")
            not_found_pages_list.append(table_counter)

driver.quit()

```
